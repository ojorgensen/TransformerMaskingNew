```python
from datasets import load_dataset

lang_keys = ["en", "es", "fr", "de", "pt"]
all_datasets = {key:[] for key in lang_keys}
n = 200
for key in lang_keys:
  dataset = iter(load_dataset("cc100", lang=key, split='train', streaming=True).shuffle(buffer_size=100000))
  # Take the first 200 examples from each
  for i in range(n):
    input = next(dataset)
    text = input['text']
    all_datasets[key].append(text)



```
